### YouTube Dislikes Dataset:
● This dataset contains information about trending YouTube videos from August 2020 to December 2021
for the USA, Canada, and Great Britain.

● This dataset contains the latest possible information about dislikes,likes,views and more which was
collected just before December 13. 

The information was collected by videos that had been trending in
the USA, Canada, and Great Britain for a year prior.

● Dataset link: https://www.kaggle.com/datasets/dmitrynikolaev/youtube-dislikes-dataset

# <font color='red'>6)Create a new column as 'published_month' using the column published_at (display the months only)</font>

```python
data['published_month']=data['published_at'].dt.month
data
```

# <font color='red'>7)Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and so on.....</font>


```python
data['published_month']=data['published_at'].dt.strftime('%B')
data['published_month']
```

# <font color='red'>8) Find the number of videos published each month and arrange the months in a decreasing order based on the video count.</font>

```python
monthpublished=data.groupby('published_month')['video_id'].count().sort_values(ascending=False)
monthpublished
```

# <font color='red'>Q10) Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.</font>


```python
top_ten=data.groupby('channel_title')['title'].count().sort_values(ascending=False).head(10)
bottom_ten=data.groupby('channel_title')['title'].count().sort_values(ascending=True).head(10)
```

```python
print("The Top 10 Channels With the Highest Number of videos are:\n",top_ten)
print("The Bottom 10 Channels With Lowest Number of Videos are:\n",bottom_ten)
```
